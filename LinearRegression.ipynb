```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv(
    filepath_or_buffer = "placement.csv"
)
dataFrame.head()
```

```python
plt.scatter(
    x = dataFrame['cgpa'],
    y = dataFrame['package']
)

plt.xlabel("CGPA")
plt.ylabel("lpa")
plt.plot()
```

```python
X = dataFrame[['cgpa']]
y = dataFrame.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.2, random_state= 2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(X = X_train, y = y_train)
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
lr.predict(X_test.iloc[1].values.reshape(1, 1))
```

```python
plt.scatter(
    x = dataFrame['cgpa'],
    y = dataFrame['package']
)

plt.plot(
    X_test,
    lr.predict(X_test),
    color = "red"
)

plt.xlabel("CGPA")
plt.ylabel("lpa")
plt.plot()
```
