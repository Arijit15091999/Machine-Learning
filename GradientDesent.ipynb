```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(
    n_samples=100,
    n_features=1,
    n_informative=1,
    n_targets = 1,
    noise=20,
    random_state=13
)

lr = LinearRegression()

lr.fit(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X), color = "red", label = "lr")
plt.legend()
plt.plot()
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
class LinearRegressionOLS:
    def __init__(self):
        self.m = 0
        self.b = 0

    def fit(self, X , y):
        X = np.ravel(X)
        y = np.ravel(y)

        numerator = np.sum((X - X.mean()) * (y - y.mean()))
        denominator = np.sum((X - X.mean()) * (X - X.mean()))

        self.coef_ = numerator / denominator
        self.intercept_ = y.mean() - self.coef_ * X.mean()

    def predict(self, X):
        return self.coef_ * X + self.intercept_
```

```python
ols = LinearRegressionOLS()

ols.fit(X, y)

print(ols.coef_, ols.intercept_)
```

```python
class SGDRegressor:
    def __init__(self, learning_rate = 0.001, epochs = 100):
        self.lr = learning_rate
        self.epochs = epochs
        self.m = 100
        self.b = -120

    def fit(self, X, y):

        X = np.ravel(X)
        y = np.ravel(y)

        for i in range(self.epochs):
            loss_slope_m = -2 * np.sum((y - self.m * np.ravel(X) - self.b) * np.ravel(X))
            loss_slope_b = -2 * np.sum(y - self.m * np.ravel(X) - self.b)

            step_size_m = self.lr * loss_slope_m
            step_size_b = self.lr * loss_slope_b

            self.m = self.m - step_size_m
            self.b = self.b - step_size_b

            print(self.m, self.b)

        self.coef_ = self.m
        self.intercept_ = self.b

    def predict(self, X):
        return self.m * np.ravel(X) + self.b

```

```python
gd = SGDRegressor(epochs=10000, learning_rate=0.001)

gd.fit(X, y)

print(gd.coef_, gd.intercept_)
```
