```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X = 6 * np.random.rand(200, 1) - 3

y = 0.8 + 0.6 * X + 0.9 * X**2 + np.random.randn(200, 1)
```

```python
plt.scatter(X, y)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
poly = PolynomialFeatures(degree=2)
X_train_transform = poly.fit_transform(X_train)
X_test_transform = poly.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_transform, y_train)
```

```python
y_pred = lr.predict(X_test_transform)
```

```python
r2_score(y_test, y_pred)
```

```python
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", label = "predicted")
plt.plot(X_train, y_train, "b.", label = "traing date")
plt.plot(X_test, y_test, "g.", label = "testing data")
plt.legend()
plt.show()
```
